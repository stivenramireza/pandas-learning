## Pandas Intermediate for Data Science

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/VinitaSilaparasetty/Coursera-Intermediate-Pandas/master/What_does_aid_to_Africa_finance_1.csv")
df.head(3)
```

### Splitting data

```python
df_new = df.copy()
```

```python
df_1 = df_new.sample(frac=0.25, random_state=0)
df_new = df_new.drop(df_1.index)
```

```python
df_2 = df_new.sample(frac=0.25, random_state=0)
df_new = df_new.drop(df_2.index)
```

```python
df_3 = df_new.sample(frac=0.25, random_state=0)
df_4 = df_new.drop(df_3.index)
```

### Detect missing values

```python
df_3.isnull().sum()
```

#### - Imputation

```python
df_3['popn'].mean()
```

```python
df_3.isnull()
```

```python
df_3['popn'].fillna(df_3['popn'].mean, inplace=True)
```

```python
df_3.isnull().sum()
```

#### - Interpolation

```python
df_1.isnull().sum()
```

```python
df_1['popn'].fillna(df_1['popn'].inteporlate(), inplace=True)
```

```python
df_1.isnull().sum()
```

### Joining

```python
df_5 = df_1.join(df_2, lsuffix='_left')
```

```python
df_5
```

### Concatenation

```python
df_6 = pd.concat([df_1, df_2], axis=0)
df_6
```

### Advanced Joins

#### Inner Join

```python
df_7 = pd.merge(df_1, df_2, on='countryc')
df_7
```

#### Full Outer Inclusive Join

```python
df_8 = pd.merge(df_1, df_2, how='outer')
df_8
```

#### Left Inclusive Join

```python
df_9 = pd.merge(df_1, df_2, how='left')
df_9
```

#### Right Inclusive Join

```python
df_10 = pd.merge(df_1, df_2, how='right')
df_10
```
